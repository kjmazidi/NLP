### Select a model

The code below was copied from the model card for this model: https://huggingface.co/distilbert/distilbert-base-uncased-finetuned-sst-2-english

Clicking on the **Use this model** button at the upper right showed the following block of code. 

In this notebook, I received a warning that I didn't have a secret HF_TOKEN. The code runs anyway. 

There may be situations when you need a Hugging Face token, such as writing to Hugging Face, or downloading private repos. To access the token, sign in to your account, go to Settings, and then get your User Access Token. 

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

The code below to get results on test data is shown below. 

```python
results = pipe(["I like brussels sprouts.", "I don't like fish."])
```

```python
for result in results:
    print(f"label: {result['label']}, score: {round(result['score'], 4)}")
```
