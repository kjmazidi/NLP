### HF pipeline

Hugging Face pipelines are provided to allow users to quickly use HF models with only very simple code. Behind the pipeline is the complex code from the API. 

The pipeline() object encapsulates all other pipelines. There are task-specific pipelines for audio, vision, NLP and more. 

Notice that after importing the pipeline, we ask for a specific sentiment analysis pipeline which will be associated with the identifier *classifier*. We didn't specify a model, so HF chose a default model. 

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
```

### Running the classifier

To classify sentiment for a few sentences, we simply input that list of sentences into the classifier. Then we iterate over the output. 

Notice that the first sample sentence got a very high positive score which seems right. The second sentence got a middling negative score which seems about right as well. 

```python
results = classifier(["We are very happy to show you the Transformers library.", "We hope you don't hate it."])
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```
